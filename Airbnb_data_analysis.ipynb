```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("compressed_data_airbnb.csv",low_memory=False)
```

```python
df.head()
```

# Checking Missing Values

```python
#To check whether it has missing value or null value or not
print(df.isnull().sum()) 
```

```python
# To check the data type contain in each column
df.info()
```

# Handling Missing Values

```python
df['last review'] = pd.to_datetime(df['last review'],errors = 'coerce')
```

```python
df.fillna({'reviews per month' : 0, 'last review' : df['last review'].min()}, inplace = True)
```

```python
df.dropna(subset = ['NAME', 'host name'], inplace = True)
```

```python
print(df.isnull().sum())
```

# Deleting unwanted columns

```python
df = df.drop(columns = ["license", "house_rules"], errors = 'ignore')
```

```python
df.head()
```

# Remove dollar signs and convert to float

```python
df['price'] = df['price'].replace('[\$,]', '',regex = True).astype(float)
df['service fee'] = df['service fee'].replace('[\$,]', '',regex = True).astype(float)
```

```python
df.head()
```

# Remove Duplicates

```python
df.drop_duplicates(inplace = True)
```

# Descriptive Statistics

```python
df.describe()
```

# Visualization :-

# 1. What is the distribution of listing prices?

```python
plt.figure(figsize = (10,6))
sns.histplot(df['price'], bins = 50 , kde = True , color = 'red')
plt.title("Distribution of Listing Price")
plt.xlabel("Price $")
plt.ylabel("Frequency")
plt.show()
```

# 2. How are different room types distributed?

```python
df['room type']
```

```python
plt.figure(figsize = (8,4))
sns.countplot( y = "room type" , data = df , color = "grey")
plt.title('Count')
plt.ylabel('Room Type')
plt.xlabel('Room Type Distribution')
plt.show()
```

# 3. How are listing distributed across different neighborhoods?

```python
plt.figure(figsize = (8,4))
sns.countplot(x = 'neighbourhood group', data = df, color='red', order = df['neighbourhood group'].value_counts().index)
plt.title('Number of Listing by Neighbourhood Group')
plt.xlabel('Count')
plt.ylabel('Neighborhood Group')
plt.show()
```

# 4. What is the relationship between price and room type?

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='room type',y='price',hue='room type',data=df,palette='Set1')
plt.title('Price vs. Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price ($)')
plt.legend(title='Room Type')
plt.show()
```

# 5. How has the number of reviews change over with time?

```python
df['last review'] = pd.to_datetime(df['last review'])
reviews_over_time = df.groupby(df['last review'].dt.to_period('M')).size()
plt.figure(figsize = (12,6))
reviews_over_time.plot(kind='line',color='red')
plt.title('Number of Reviews Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Reviews')
plt.show()
```
